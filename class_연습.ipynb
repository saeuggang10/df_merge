# class

```python
import pandas as pd
import os
```

```python
class DF_temp:

    def __init__(self, path_temp, len_temp, header_temp, skip_temp, usecol_temp):
        self.path = path_temp
        self.len_temp = len_temp
        self.header_temp = header_temp
        self.skip_temp = skip_temp
        self.usecol_temp = usecol_temp
        self.df=pd.DataFrame()

        file = os.listdir(self.path)
        for i in range(0,self.len_temp):
            df1 = pd.read_excel(self.path + file[i], header=self.header_temp,
                                skipfooter=self.skip_temp, usecols=self.usecol_temp)
            self.df = pd.concat([self.df, df1], ignore_index=True)
            
    def read(self):
        return self.df
    
    def filter(self, val_name, lis_temp, bool_temp=True):
        con = self.df[val_name].isin(lis_temp)
        return self.df[con == bool_temp]
    
    def add(self, val_name, contents):
        self.df[val_name] = contents
        return self.df
    
    def save(self, file_name, index_temp=False):
        self.df.to_excel(self.path + file_name + '.xlsx', index=index_temp)
```
